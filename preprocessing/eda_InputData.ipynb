```python
# creating a funtion to check null values from all columns and store column names in a list
def check_null(colums_list):
    col_list_null_values = [] # creating an empty list in which column name will be stored to fill null values
    for col_name in colums_list:
        if df[col_name].isnull().sum() > 0:
            print(f"{col_name} has {df[col_name].isnull().sum()} null values")
            col_list_null_values.append(col_name)
        else:
            print(f"{col_name} has no null values")
    return col_list_null_values

# check_null(colums_list)
```

```python
# checking how many columns need to be dropped or fill the null values
total_cols = check_null(colums_list)
```

```python
colums_list = ['Agent_Name', 'Officer_Name', 'SITE', 'Source', 'EDate',
       'ACPT_Category', 'ACPT_Type', 'ACPT_Subtype', 'ACPT_Reason',
       'Responsiblity','Update_By', 'Survey_date', 'Group_Name',
       'CWC_Cat', 'CWC_Lvl1', 'CWC_Lvl2', 'CWC_Lvl3',
       'Segment', 'Sub_Segment', 'Product', 'Action_Status', 'ACPT_Secondary',
       'SCIM_Usage', 'Referance_Type', 'Audit Date', 'Tag',
       'Sub Tag']
```

```python
# funtion to replace NaN values with the previous record value (forward fill) or backward fill
def fill_null(df, colums_list):
    for col_name in colums_list:
        df[col_name].fillna(method='ffill', inplace=True)
        df[col_name].fillna(method='bfill', inplace=True)
    return df
```

```python
# calling the fill_null funtion to fill the null values
fill_null(df, colums_list)
```

```python
# Load the data
# df_filter = pd.read_excel("preprocessed_filter.xlsx")

# Convert 'EDate' to datetime format, coercing errors
df['EDate'] = pd.to_datetime(df['EDate'], errors='coerce')

# Define the date to be removed
drop_date = pd.to_datetime('1970-01-01').date()  # Use .date() to keep only the date part

# Drop rows where 'EDate' is equal to '1/1/1970'
df_cleaned = df[df['EDate'].dt.date != drop_date]

# Convert 'EDate' to date format
df_cleaned['EDate'] = df_cleaned['EDate'].dt.date

# Save the cleaned DataFrame to an Excel file
df_cleaned.to_excel('preprocessed_data1.xlsx', index=False)
```
